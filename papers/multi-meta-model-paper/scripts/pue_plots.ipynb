```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os

print("current path is ", os.getcwd())
```

```python
def get_pue_evolution_2007_2023():    
    # Data
    years = [2007, 2009, 2011, 2013, 2015, 2017, 2019, 2021, 2023]
    pue_values = [2.5, 1.98, 1.65, 1.58, 1.67, 1.59, 1.57, 1.55, 1.58]
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(years, pue_values, marker='o', color='darkred', linestyle='-')
    plt.title('Average PUE between 2007-2023', size=14)
    plt.xlabel('Year', size=18)
    plt.ylabel('Average PUE', size=18)
    plt.grid(False)
    plt.ylim(0.8, 2.6)
    plt.xticks(years)  # Rotate years for better readability
    plt.tight_layout()  # Adjust layout to make room for the rotated x-axis labels
    
    # Add a discrete green shadow between 1.3 and 1.4 with the text
    plt.axhspan(1.3, 1.4, xmin=0.045, xmax=0.955, color='#00a3cc', alpha=0.5, lw=0)
    plt.text(2015, 1.325, 'PUE target by 2030 (Climate Neutral Data Centre Pact)', horizontalalignment='center', color='white', fontsize=12)
    
    # Make the ticks bigger
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14, ticks=[1.0, 1.3, 1.4, 1.8, 2.2, 2.6])
    
    # Add a horizontal dotted line at 1.0 with label "optimal"
    plt.axhline(y=1.0, xmin=0.045, xmax=0.955, color='green', linestyle='--', linewidth=2)
    plt.text(2008, 1.02, 'Optimal', color='green', fontsize=18)
    
    # Save the plot as a PDF
    plt.savefig('./plots/pue_evolution_2007_2023.pdf')
    plt.show()
    

get_pue_evolution_2007_2023()
```
